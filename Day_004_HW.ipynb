```python
(df.style
     .format('{:.1f}', subset='AMT_INCOME_TOTAL')
     .set_caption('★五顏六色の數據集☆')
     .hide_index()
     .bar('AMT_INCOME_TOTAL', vmin=0)
     .highlight_max('AMT_CREDIT')
     .background_gradient('Greens', 
                          subset='AMT_ANNUITY')
     .highlight_null()
)
```
